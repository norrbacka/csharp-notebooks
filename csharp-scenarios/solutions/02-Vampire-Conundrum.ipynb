# Vampire Registry Generation


```csharp
using System;
using System.Collections.Generic;

private static Random random = new Random();
private static Dictionary<int, string> vampireTypes = new Dictionary<int, string>
{
    {-6, "Disfigured Vampire"},
    {-5, "Insane Vampire"},
    {-4, "Soul Vampire"},
    {-3, "Psi-Vampire"},
    {-2, "Emphatic Vampire"},
    {-1, "Half-Vampire"},
    {0, "Regular Human"},
    {1, "Blood Drinkers Vampire"},
    {2, "Genetic Vampire"},
    {3, "Moonbane Vampire"},
    {4, "Psychic Vampire"},
    {5, "Vampyre"},
    {6, "Origin Lord Vampire"}
};

private static double GenerateGaussian(double mean, double stdDev)
{
    // Using Box-Muller transform to generate Gaussian distributed numbers
    double u1 = 1.0 - random.NextDouble(); // Uniform(0,1] random doubles
    double u2 = 1.0 - random.NextDouble();
    double randStdNormal = Math.Sqrt(-2.0 * Math.Log(u1)) * Math.Sin(2.0 * Math.PI * u2);
    return mean + stdDev * randStdNormal;
}

public static string GetRandomVampire()
{
    double mean = 0; // Centered on "Regular Human"
    double stdDev = 2; // Adjust standard deviation as needed

    int typeIndex;
    do
    {
        typeIndex = (int)Math.Round(GenerateGaussian(mean, stdDev));
    } while (!vampireTypes.ContainsKey(typeIndex));

    return vampireTypes[typeIndex];
}

List<string> registry = new List<string>();
for (int i = 0; i < 1000; i++) // Generate 100 entries
{
    registry.Add(GetRandomVampire());
}

foreach (var type in registry)
{
    Console.WriteLine(type);
}
```
